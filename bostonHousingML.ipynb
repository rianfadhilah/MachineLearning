<b> Prepare data

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
df.columns
```

```python
df_y= df['medv']
df_x= df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
```

```python
df_x.describe()
```

<b>Initialize the linear regression model

```python
req = linear_model.LinearRegression()
```

<b>Split data 67% training and 33% testing data

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size= 0.33, random_state=42)
```

<b>normalize data

```python
x_train = preprocessing.scale(x_train)
x_test = preprocessing.scale(x_test)
```

<b>Training the data

```python
req.fit(x_train, y_train)
```

```python
print(req.coef_)
```

<b>Testing the data

```python
x_pred = req.predict(x_test)
print(x_pred)
```

```python
print(y_test)
```

<b> Check the model Accuracy using mean squared error (MSE)

Mewakili perbedaan antara nilai asli dan prediksi yang diekstraksi dengan mengkuadratkan perbedaan rata-rata pada kumpulan data.

```python
print("Accuracy using MSE: ", np.mean((x_pred - y_test)**2))
```

<b> Check the model Accuracy using mean squared error (MSE) and sklearn.metrics

```python
from sklearn.metrics import mean_squared_error
print("Accuracy using MSE dari sklearn: ",mean_squared_error(y_test,x_pred))
```

<b>Check the model Accuracy using mean_absolute_error(MAE)

Mewakili perbedaan antara nilai asli dan nilai prediksi yang diekstraksi dengan rata-rata perbedaan absolut pada kumpulan data.

```python
from sklearn.metrics import mean_absolute_error
print("Accuracy using MAE",mean_absolute_error(y_test,x_pred))
```

<b> Accuracy using r2_score

Mewakili koefisien seberapa cocok nilai-nilai tersebut dibandingkan dengan nilai aslinya. Nilai dari 0 sampai 1 diartikan sebagai persentase. Semakin tinggi nilainya, semakin baik modelnya.

```python
from sklearn.metrics import r2_score
```

```python
accuracy_score = req.score(x_test, y_test)
```

```python
print(f'Accuracy using r2 score: {accuracy_score * 100:.2f}')
```
